```python
def sigmoid(z):
    z = np.clip(z, -500, 500)
    z = 1 / (1 + np.exp(-z))
    return z

def dense(a_in, W, b):
    units = W.shape[1]
    a_out = np.zeros(units)
    for j in range(units):
        # selecting all rows and jth column
        w = W[:,j]
        z = np.dot(w,a_in) + b[j]
        a_out[j] = sigmoid(z)

    return a_out

# take the output for each layer
def Sequential(x,W1,b1,W2,b2):
    a1 = dense(x, W1, b1)
    a2 = dense(a1, W2, b2)

    return a2
    
#  we need two weights 1 for each
W1 = np.array( [[-8.93,  0.29, 12.9 ], [-0.1,  -7.32, 10.81]] )
print("W1 shape : ", W1.shape)
b1 = np.array( [-9.82, -9.28,  0.96] )

# output layer has 1 neuron but needs previous layers weights size
W2 = np.array( [[-31.18], [-27.59], [-32.56]] )
print("W2 shape : ", W2.shape)
b2 = np.array( [15.41] )
```

```python
def predict(x, W1, b1, W2, b2):

    # run through each sample and get the output
    m = x.shape[0]
    p = np.zeros((m,1))
    for i in range(m):
        output = Sequential(x[i], W1, b1, W2, b2)
        
        # Explicitly extract the scalar value for the prediction
        p[i, 0] = output[0]

    return p

output = predict(x, W1, b1, W2, b2)
print(output.shape)
```
